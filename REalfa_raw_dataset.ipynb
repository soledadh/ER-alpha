```python
# Import libraries
import pandas as pd
from chembl_webresource_client.new_client import new_client
```

```python
# Target search

target = new_client.target
target_query = target.search('estrogen receptor alpha')
```

```python
targets = pd.DataFrame.from_dict(target_query)
targets
```

```python
target_realpha = targets.target_chembl_id[2]
target_realpha
```

```python
activity = new_client.activity
res = activity.filter(target_chembl_id=target_realpha).filter(standard_type="EC50")
```

```python
df_EC50 = pd.DataFrame.from_dict(res)
df_EC50
```

```python
df_EC50.columns
```

```python
df_EC50.to_csv('ERalpha_01_bioactivity_data_raw(Chembl).csv', index=False)
```
